```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("/kaggle/input/smartphone-sale-dataset/Sales.csv")
```

HANDLING NULL VALUES

```python
df['Rating']=df['Rating'].fillna(df['Rating'].mean())
```

```python
df['Memory']=df['Memory'].fillna(df['Memory'].mode())
```

```python
df['Storage']=df['Storage'].fillna(df['Storage'].mode())
```

VISULAIZATION

```python
 label=LabelEncoder()
df['Brands']=label.fit_transform(df['Brands'])
df['Models']=label.fit_transform(df['Models'])
df['Colors']=label.fit_transform(df['Colors'])
df['Memory']=label.fit_transform(df['Memory'])
df['Storage']=label.fit_transform(df['Storage'])
df['Camera']=label.fit_transform(df['Camera'])
df['Rating']=label.fit_transform(df['Rating'])
df['Mobile']=label.fit_transform(df['Mobile'])
```

```python
df.dtypes
```

```python

plt.figure(figsize=(20,16))
sns.heatmap(df.corr(),annot=True)
plt.title('HEAT MAP')
```

```python
#Linear Regression
X=df.drop(columns='Selling Price')
y=df['Selling Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
lin_model=LinearRegression()
lin_model.fit(X_train,y_train)
y_pred=lin_model.predict(X_test)

```

```python
me=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(me)
```

```python
print("Mean Squared Error: ",me)
print("Root Mean Squared Error: ",rmse)
```
